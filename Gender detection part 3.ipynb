```python
from keras.models import load_model
import numpy as np
import cv2

model = load_model('GenderDetecter1.h5')


face_clf = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

source = cv2.VideoCapture(0) #0 means open web cam of coputer

labels_dict={0:'Male',1: 'Female'}
color_dict={0:(255,0,0),1:(0,255,0)}
```

```python


#loading the image and converting it to greyscale
image = cv2.imread('images\img6.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = face_clf.detectMultiScale(gray, 1.3, 5)
Text = ('Found {0} Persons !!'.format(len(faces)))

for (x, y, w, h) in faces:
    face_img = gray[y:y+w, x:x+h]
    
    #resizing and reshaping the image
    resized = cv2.resize(face_img, (100, 100))
    normalized = resized / 255.0
    reshaped = np.reshape(normalized, (1, 100, 100, 1))
    result = model.predict(reshaped)
    
    label = np.argmax(result, axis=1)[0]
    
    cv2.rectangle(image, (x, y), (x+w, y+h), color_dict[label], 2)  # color_dict[label] means if result is 0 then its give red color
    cv2.rectangle(image, (x, y-40), (x+w, y), color_dict[label], -1)
    cv2.putText(image, labels_dict[label], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    
    cv2.rectangle(image,pt1=(220,399),pt2=(10,350),color=(255,0,0),thickness=2)
    cv2.putText(image, Text,(20,380),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
while(True):
    
    ret,img= source.read()
    gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_clf.detectMultiScale(gray,1.3,5) #here last values are face detecting size value
    Text = ('Found {0} Persons !!'.format(len(faces)))
    
    for (x,y,w,h) in faces:
        
        face_img = gray[y:y+w,x:x+h]
        resized=cv2.resize(face_img,(100,100))
        normalized = resized/255.0
        reshaped= np.reshape(normalized,(1,100,100,1))
        result = model.predict(reshaped)
        
        label = np.argmax(result,axis=1)[0]  #thi is used  to find result is 0 or 1
        
        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)#here colordic[label] means if result is 0 then its give red color
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
        cv2.rectangle(img,pt1=(220,445),pt2=(10,395),color=(255,0,0),thickness=2)
        cv2.putText(img, Text,(30,420),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)
        
    cv2.imshow('Gender detection',img)
    key=cv2.waitKey(1)
    
    if(key==27):
        break
        
cv2.destroyAllwindows()
source.release()
```
